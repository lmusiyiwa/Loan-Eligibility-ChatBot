```python
# üíé Discovery LoanBot v31 ‚Äî Discovery's Digital Loan System and Credit Education 

import time, datetime, random, sys

# ANSI Colors
BEIGE = "\033[38;5;230m"
RESET = "\033[0m"

# Utility Functions 
def slow_print(text, delay=0.03):
    """Simulate typing effect for bot messages in beige."""
    for char in f"{BEIGE}{text}{RESET}":
        sys.stdout.write(char)
        sys.stdout.flush()
        time.sleep(delay)
    print()

def pause(text, delay=0.7):
    slow_print(text)
    time.sleep(delay)

def ask_nonempty(prompt):
    while True:
        try:
            ans = input(prompt).strip()
        except (EOFError, KeyboardInterrupt):
            slow_print("\n‚ùó Input interrupted. Exiting.")
            raise SystemExit
        if ans == "":
            slow_print("‚ö†Ô∏è Please type a response (not empty).")
            continue
        return ans

def ask_int(prompt):
    while True:
        s = ask_nonempty(prompt)
        try:
            return int(s)
        except ValueError:
            slow_print("‚ö†Ô∏è Please enter a whole number (e.g., 30).")

def ask_float(prompt):
    while True:
        s = ask_nonempty(prompt)
        try:
            return float(s)
        except ValueError:
            slow_print("‚ö†Ô∏è Please enter a number (e.g., 6000).")

def ask_yesno(prompt):
    while True:
        s = ask_nonempty(prompt).lower()
        if s in ("yes", "y"): return "yes"
        if s in ("no", "n"): return "no"
        slow_print("‚ö†Ô∏è Please answer 'yes' or 'no' (or 'y'/'n').")

def loading_bar(task, duration=2.5):
    slow_print(f"\nüîç {task}")
    for i in range(10):
        bar = "‚ñ∞" * (i + 1) + "‚ñ±" * (9 - i)
        sys.stdout.write(f"\r{BEIGE}{bar} {(i+1)*10}%{RESET}")
        sys.stdout.flush()
        time.sleep(duration / 10)
    sys.stdout.write(f"\r{BEIGE}‚ñ∞‚ñ∞‚ñ∞‚ñ∞‚ñ∞‚ñ∞‚ñ∞‚ñ∞‚ñ∞‚ñ∞ 100% ‚úÖ{RESET}\n\n")
    sys.stdout.flush()

# Greeting 
hour = datetime.datetime.now().hour
greeting = "Good morning" if hour < 12 else "Good afternoon" if hour < 18 else "Good evening"

pause(f"\nü§ñ {greeting}! Welcome to üíé Discovery DigiBot v3.2 üíé", 1)
pause("Your smart AI loan and credit advisor.\n")

# Main Menu
while True:
    pause("üìã Main Menu:")
    slow_print("1Ô∏è‚É£ Check loan eligibility")
    slow_print("2Ô∏è‚É£ Learn about credit score")
    slow_print("3Ô∏è‚É£ Exit\n")

    choice = ask_nonempty("Choose an option (1-3): ")

    # 1Ô∏è‚É£ Loan Eligibility 
    if choice == "1":
        pause("\nüí∞ Let's check your Discovery loan eligibility!\n", 1)
        customer = {}

        age = ask_int("Please enter your age: ")
        real_estate = ask_yesno("Do you own real estate? (yes/no): ")

        customer.update({"Age": age, "Real_Estate": real_estate})
        result = ""
        credit_score = random.randint(450, 850)

        # --- Core Logic (Auto-approve real estate owners) ---
        if real_estate == "yes":
            result = "üè† Loan Approved ‚úÖ (Property-backed loan)"
        else:
            income = ask_float("What is your monthly income (in Rands)? ")
            customer["Income"] = income

            if age > 40:
                if income >= 7000:
                    result = "üí∞ Loan Approved ‚úÖ (Good income stability)"
                else:
                    result = "üö´ Loan Declined ‚Äî Insufficient financial standing ‚ùå"
            else:
                if income > 5000:
                    result = "üí∞ Loan Accepted ‚úÖ (Meets minimum criteria)"
                else:
                    education = ask_nonempty("What's your education level? (High/Secondary/Special): ").lower()
                    customer["Education"] = education
                    if education == "high":
                        result = "üéì High education ‚Äî Potential for review at branch."
                    elif education == "secondary":
                        result = "üéì Secondary education ‚Äî May qualify for smaller educational loan."
                    elif education == "special":
                        result = "üéì Special education ‚Äî Student/specialist loan possible."
                    else:
                        result = "üö´ Loan Declined ‚Äî Does not meet criteria ‚ùå"

        # Credit score simulation 
        loading_bar("Evaluating your Discovery credit profile...", 3)
        pause(f"üìà Estimated Credit Score: {credit_score}/850")

        if credit_score >= 720:
            tier = "üåü Excellent"
        elif credit_score >= 660:
            tier = "‚úÖ Good"
        elif credit_score >= 580:
            tier = "‚ö†Ô∏è Fair"
        else:
            tier = "‚ùå Poor"

        pause("\nüìä Summary of your responses:")
        for k, v in customer.items():
            slow_print(f"   - {k}: {v}")
        pause(f"   - Credit Score: {credit_score} ({tier})")
        pause(f"\n‚ú® Result: {result}\n", 1)

    # --- 2Ô∏è‚É£ Learn About Credit Score ---
    elif choice == "2":
        pause("\nüí° Let's estimate your credit score and help you improve it.\n")
        name = ask_nonempty("What's your first name? ")

        income = ask_float("üí∞ What is your monthly income (in Rands)? ")
        debts = ask_float("üí≥ Total outstanding debts (in Rands)? ")
        missed = ask_int("‚ùå How many payments have you missed in the last 12 months? ")
        usage = ask_int("üìä What % of your available credit are you using? (0-100): ")

        # Simple credit score model
        base_score = 600
        score = base_score + (income / 1000) - (debts / 2000) - (missed * 50) - (usage / 2)
        score = max(200, min(int(score), 850))

        pause(f"\nüìä {name}, your estimated credit score is: {score}")

        if score < 580:
            pause("\n‚ö†Ô∏è Your credit score is LOW. But don‚Äôt worry, Discovery LoanBot is here to help!\n")
            slow_print("Here are some FREE resources:")
            slow_print("üí° Free Financial Advisor: National Credit Regulator (NCR) ‚Äì Financial Education Programme")
            slow_print("üí° Free Debt Counselling: Debt Counsellors Association of South Africa (DCASA) helpline")
            slow_print("üí° Free Therapy Support: South African Depression and Anxiety Group (SADAG) helpline\n")

        # Tips
        slow_print("Expert tips to improve your credit score:")
        slow_print("‚úÖ Pay all bills on time ‚Äî even small payments count.")
        slow_print("‚úÖ Reduce total debt gradually.")
        slow_print("‚úÖ Keep credit usage below 30%.")
        slow_print("‚úÖ Avoid too many new credit accounts.")
        slow_print("‚úÖ Monitor your credit report for errors.\n")

        # Quiz
        pause("üí° Quick question:")
        slow_print("Which action helps improve your credit score fastest?")
        slow_print("1Ô∏è‚É£ Miss a payment occasionally")
        slow_print("2Ô∏è‚É£ Reduce outstanding debt")
        slow_print("3Ô∏è‚É£ Open many new credit accounts")
        slow_print("4Ô∏è‚É£ Ignore your credit report")

        ans = ask_nonempty("Your choice (1-4): ")
        if ans == "2":
            pause("‚úÖ Correct! Reducing debt improves your credit score fastest.")
        else:
            pause("‚ùå Not quite. Reducing debt is the fastest way to boost your score.")

        pause(f"\nüëè Well done, {name}! Keep these habits consistent to grow your financial health. üíº\n")

    # --- 3Ô∏è‚É£ Exit ---
    elif choice == "3":
        pause("\nüëã Thank you for using Discovery LoanBot v3! Stay financially empowered.\n")
        break

    else:
        pause("‚ö†Ô∏è Invalid option. Please enter 1, 2, or 3.\n")
```
